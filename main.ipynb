```python
from router import decide_route_and_build_prompt
from models import call_local_model, call_cloud_model
```

```python
fake_session = {
    "username": "test_user",
    "traits": {
        "persona_description": "A CS student who prefers detailed explanations.",
        "tone_preferences": "detailed",
        "expertise_level": "intermediate",
        "preferred_language": "en",
        "response_style": "use examples"
    }
}
```

```python
def test_chat(fake_session, user_input):
    traits = fake_session["traits"]

    routing = decide_route_and_build_prompt(traits, user_input)
    route = routing["route"]
    final_prompt = routing["final_prompt"]

    print(f"\nUser request: {user_input}")
    print(f"[Router chose: {route}]")

    if route == "LOCAL_SMALL_MODEL":
        response = call_local_model(final_prompt)
    else:
        response = call_cloud_model(final_prompt)

    print("\nAssistant:", response)
```

```python
tests = [
    "What is a hashmap?",
    "Write an essay comparing capitalism and socialism.",
    "Explain neural networks with code.",
    "Tell me a joke.",
]

for t in tests:
    test_chat(fake_session, t)
    print("-" * 60)
```
