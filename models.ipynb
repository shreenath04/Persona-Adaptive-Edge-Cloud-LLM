```python
import ollama
```

```python
LOCAL_MODEL_NAME = "gemma3:4b"
CLOUD_MODEL_NAME = "llama3.1:8b"
```

```python
def call_local_model(prompt: str) -> str:

    response = ollama.chat(
        model = LOCAL_MODEL_NAME,
        messages = [
            {"role": "system", "content": "You are a localhelpful assistant."},
            {"role": "user", "content": prompt}
        ]
    )
    return response["message"]["content"].strip()
```

```python
def call_cloud_model(prompt: str) -> str:

    response = ollama.chat(
        model = CLOUD_MODEL_NAME,
        messages = [
            {"role": "system", "content": "You are a powerful AI assistant."},
            {"role": "user", "content": prompt}
        ]
    )
    return response["message"]["content"].strip()
```

```python
print(call_local_model("Describe yoursleves in few words"))
```

```python
print(call_cloud_model("Describe yoursleves in few words"))
```

```python
print(call_local_model("describe how load balancing works in system design"))
```

```python
print(call_cloud_model("Describe how load balancing works in system design"))
```
